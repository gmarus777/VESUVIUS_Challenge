## Sanity check
### Let's simulate how the model will traverse the image in patches during training
- if increasing the number of pathces the validation region will become apparent (by default it is around letter $\rho$ in the first scroll

<span style="color:red">**Note**: *for matplotlib.plt patching need to swap x and y*</span>.

```python
fig, ax = plt.subplots()
ax.imshow(dataset.data_train.label)
BUFFER = 15
for i in range(10):
    k = np.random.randint(1,len(dataset.data_train.pixels))
    x,y = dataset.data_train.pixels[k] # for patches need to swap x and y
    patch = patches.Rectangle([y - BUFFER, x - BUFFER], 2 * BUFFER, 2 * BUFFER, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(patch)
plt.show()
```
